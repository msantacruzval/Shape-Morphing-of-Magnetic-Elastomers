```python
Ly = 3 # mm, width of the unit cell
T = 5 # s
```

```python
# Example time steps
t_steps = [t2, t4, t8, t19]

skip = [39,20,15,8]
# Create the figure and axes
fig, ax1 = plt.subplots(figsize=(6, 2.5), layout='constrained')


j = 0
for i in t_steps:
    # Plotting every fifth point
    ax1.plot(i['Time'][::skip[j]]/T, i['dy'][::skip[j]]/Ly, '-o', label=str(int(T/i['dt'][0])))
    j += 1

# Set the labels with proper LaTeX formatting
ax1.set_xlabel(r'$\frac{t}{T}$', fontsize=20)
ax1.set_ylabel(r'$\alpha_y(t)$', rotation='horizontal', va='center', labelpad=30)

# Add legend
ax1.legend(title='Num. Steps')

ax1.set_xticks(np.arange(0,1.1,0.1))
ax1.set_yticks(np.linspace(0, 0.3, 6))

# plt.savefig(f'{wd}/Time_Step_Convergence.svg', dpi=300)
# plt.savefig(f'{wd}/Time_Step_Convergence.png', dpi=300)

plt.show()
```

```python
# Example time steps
t_steps = [t2, t4, t8, t19]

skip = [39,20,15,8]

fig, ax = plt.subplots(2, 1, sharex=True, figsize=(7, 4), layout='constrained')


bright_palette = [
    "#0072B2",  # Blue
    "#D55E00",  # Orange
    "#F0E442",  # Yellow
    "#009E73",  # Green
    "#CC79A7",  # Pink
    "#56B4E9",  # Light Blue
    "#E69F00",  # Golden Orange
    "#8E44AD"   # Purple (additional color)
]


j = 0
for i in t_steps:
    # Plotting every fifth point
    ax[0].plot(i['Time'][::skip[j]]/T, i['dy'][::skip[j]]/Ly, '-o', label=str(int(T/i['dt'][0])), c=bright_palette[j])
    j += 1


ax[0].set_ylabel(r'$\alpha_y(t)$', rotation='horizontal', va='center', labelpad=30)


ax[0].set_yticks(np.linspace(0, 0.3, 6))


# Example time steps
m_steps = [m2, m4, m8, m19]

skip = [3,6,8,12]


j = 0
for i in m_steps:
    # Plotting every fifth point
    ax[1].plot(i['Time'][::skip[j]]/T, i['dy'][::skip[j]]/Ly, '-o', label=str(int(Ly/i['ds'][0])), c=bright_palette[j+4])
    j += 1

# Set the labels with proper LaTeX formatting
ax[1].set_xlabel(r'$\frac{t}{T}$', fontsize=20)
ax[1].set_ylabel(r'$\alpha_y(t)$', rotation='horizontal', va='center', labelpad=30)

# Add legend
ax[0].legend(title='Num. Steps', bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0.1)
ax[1].legend(title=r'$L_y$ Edge Elements', bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0.1)

ax[1].set_xticks(np.arange(0,1.1,0.1))
ax[1].set_yticks(np.linspace(0, 0.3, 6))

plt.savefig(f'{wd}/Combined_Convergence.svg', dpi=300)
plt.savefig(f'{wd}/Combined_Convergence.png', dpi=300)

plt.show()
```
