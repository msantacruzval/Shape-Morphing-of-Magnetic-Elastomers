```python
f_constant = pd.read_csv(f'{fwd}/Constant/Constant.csv', header=None, names=['Time', 'Bext', 'Vertical Disp', 'Horizontal Disp'])
f_neg_constant = pd.read_csv(f'{fwd}/Negative_Constant/Negative_Constant.csv', header=None, names=['Time', 'Bext', 'Vertical Disp', 'Horizontal Disp'])
f_increasing = pd.read_csv(f'{fwd}/Increasing/Increasing.csv', header=None, names=['Time', 'Bext', 'Vertical Disp', 'Horizontal Disp'])
f_sinusoidal = pd.read_csv(f'{fwd}/Sinusoidal/Sinusoidal.csv', header=None, names=['Time', 'Bext', 'Vertical Disp', 'Horizontal Disp'])
```

```python
f_constant
```

```python
# The constants for the unit cell

T = 5 # s
Lm = 10 # mm Length of member
```

## Final Graph

```python
dfs = [constant, neg_constant, increasing, sinusoidal]
f_dfs = [f_constant, f_neg_constant, f_increasing, f_sinusoidal]

field = ['Constant', r'(-ve) Constant', 'Increasing', 'Sinusoidal']

fig, ax = plt.subplots(2, 1, sharex=True, figsize=(7, 3.5), layout='constrained')


j = 0
for i,k in zip(dfs, f_dfs):
    # Plotting every fifth point
    ax[0].plot(k['Time']/T, k['Bext'], label=field[j])
    ax[1].plot(i['Time']/T, i['Member Disp']/Lm)
    j += 1

# Add legend
ax[0].legend(title='Field', bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0.1)


ax[1].set_xlabel(r'$\frac{t}{T}$', fontsize=20)

# Set the labels with proper LaTeX formatting
ax[0].set_ylabel(r'$B(t)$', rotation='horizontal', va='center', labelpad=42)
ax[1].set_ylabel(r'$\beta_x (t)$', rotation='horizontal', va='center', labelpad=30)

ax[1].set_xticks(np.linspace(0, 1, 10))

# plt.savefig(f'{wd}/Member_Response.svg', dpi=300)
# plt.savefig(f'{wd}/Member_Response.png', dpi=300)

plt.show()
```
